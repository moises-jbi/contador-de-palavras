### 1- Aquisição dos Dados

```python
import urllib.request  

texto = []

f = open('O Guia do Scrum.txt','r',encoding='utf-8')
for line in f:
    texto.append(line)
    
for i in range(len(texto)):
    texto[i] = texto[i].lower()
    texto[i] = texto[i].replace('\n','')
    texto[i] = texto[i].replace('.','')
    texto[i] = texto[i].replace(',','')
    texto[i] = texto[i].replace('(','')
    texto[i] = texto[i].replace(')','')
    texto[i] = texto[i].replace('?','')
    texto[i] = texto[i].replace('\'','')
```

### 2 - Construção do Dicionário

```python
mydict = {}                            

for line in texto:

  line = line.lower()                  
  words = line.split()                
#  print(words)

  for word in words:                   
    if word not in mydict.keys():        
      mydict[word] = 1                    
    else:                                  
      mydict[word] =  mydict[word] + 1    

print(mydict)
```

### 3 - Exibição dos Resultados

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.DataFrame(mydict.items(),  columns=['word', 'count']).sort_values('count',ascending=False)
df = df[df['count'] > 4] 
plt.figure(figsize=(24,10))
mpl.style.use(['fivethirtyeight'])
sns.barplot(x=df.word,y=df['count'])
plt.xticks(rotation=90) 

plt.show()
```
